```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv("C:/Users/ARSHIYA/Downloads/Unemployment in India.csv")
print(df.head())
print(df.info())
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
df['date'] = pd.to_datetime(df['date'], dayfirst=True)
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
print(df.columns.tolist())
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='date', y='estimated_unemployment_rate_(%)', data=df)
plt.title('Unemployment Rate Over Time in India')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.lineplot(x='date', y='estimated_employed', hue='region', data=df)
plt.xticks(rotation=90)
plt.title('Unemployment Rate by State')
plt.tight_layout()
plt.show()
```
